```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

plt.scatter(
    df['sepal length (cm)'],
    df['sepal width (cm)']
)

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Iris Dataset')

plt.show()
```
